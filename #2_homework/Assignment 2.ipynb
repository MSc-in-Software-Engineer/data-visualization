__7) Describe the outliers by box plot: Write a a loop for the object columns of df3 whose number of unique values is less than 10.  <font color='purple'> 15 points </font>__

```python
plt.figure(figsize=(12,8))
loop_results = df3.loc[:, df3.apply(lambda x: x.nunique()) <= 10]
loop_results.plot()
```

__8) Describe the correlation of df3: <font color='purple'> 10 points </font>__

```python
kendall_corr = df3.corr(method="kendall")
pearson_corr = df3.corr(method="pearson")

print(f"\n\nKendal Correlation;\n\n{kendall_corr}\n")

sns.heatmap(kendall_corr.corr(), annot=True)
plt.show()

print(f"\n\nPearson_corr Correlation;\n\n{pearson_corr}\n")

sns.heatmap(pearson_corr.corr(), annot=True)
plt.show()
```
