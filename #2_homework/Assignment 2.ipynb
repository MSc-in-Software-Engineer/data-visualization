__10) Describe the address # of rooms, # of bathrooms, distance for the highest Price house?: <font color='purple'> 10 points </font>__

```python
max_by_rows = df3.max()
min_by_rows = df3.min()


def describe_price_parser(results_df):
    price: float = results_df["Price"]
    results_row: dict = {
        "address": results_df["Address"],
        "rooms": results_df["Rooms"],
        "bathroom": results_df["Bathroom"],
        "distance": results_df["Distance"],
    }

    return results_row, price


max_price_row, max_price = describe_price_parser(results_df=max_by_rows)
print(f"Highest House, Price: {max_price} $\nRow: {max_price_row}\n\n")

min_price_row, min_price = describe_price_parser(results_df=min_by_rows)
print(f"Lowest House, Price: {min_price} $\nRow: {min_price_row}\n\n")
```

__11) Is there any significant relation between Regionname and Price? - <font color='purple'> 20 points </font>__

```python
from scipy.stats import pearsonr

ROUND_DEFINE: int = 4
SIGNIFICANT_RELATION_COEFFICIENT: float = 0.05

df3["Regionname"] = df3["Regionname"].astype("category").cat.codes
df3["Price"] = df3["Price"].astype("category").cat.codes

corr_coefficient, p_value = pearsonr(df3["Regionname"], df3["Price"])

# The first value in the output displays the correlation coefficient (0.0549)
print(f"correlation coefficient: {round(corr_coefficient, ROUND_DEFINE)}")

# The second value displays the p-value (0.0) associated with this correlation coefficient.
print(f"p-value: {round(p_value, ROUND_DEFINE)}")

# Since the p-value is not less than Î± = 0.05, we would conclude that the correlation between
# points and assists is not statistically significant.
if p_value <= SIGNIFICANT_RELATION_COEFFICIENT:
    print("\n\nNot significant relation between Regionname and Price!..")

if p_value >= SIGNIFICANT_RELATION_COEFFICIENT:
    print("\n\nThere is a significant relationship between Region Name and Price!..")
```
